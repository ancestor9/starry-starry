```python
# prompt: 품목의 값이 카스캔맥주, 짜파게티, 옛날국수, 페리오치약 인 것만 데이터로 추출해줘

import pandas as pd

df = pd.read_excel('/content/vlookup 함수 예제파일.xlsx', header=2, usecols='A:H')

# 품목 열의 값이 '카스캔맥주', '짜파게티', '옛날국수', '페리오치약'인 행만 추출
target_items = ['카스캔맥주', '짜파게티', '옛날국수', '페리오치약']
filtered_df = df[df['품목'].isin(target_items)]

# 결과 출력
filtered_df
```

```python
# prompt: filtered_df에 새로운 컬럼 '음식'을 만들고 '분류' 컬럼'의 내용이 식료품이 경우에만  ' 맛있다', 식료품이 아닌 나머지는 '안 맛있다'라고 해줘

filtered_df['음식'] = filtered_df['분류'].apply(lambda x: '맛있다' if x == '식료품' else '안 맛있다')
filtered_df
```

```python
# prompt: 구글에서 'Nocode"를 검색하여 해당 기사 사이트 10개만 가져오는 코드

!pip install google-search-results

from serpapi import GoogleSearch

# Replace with your actual SerpApi key
serpapi_key = "YOUR_SERPAPI_KEY"  # Get your key here: https://serpapi.com/manage-api-key

params = {
    "engine": "google",
    "q": "Nocode",
    "num": 10,
    "api_key": serpapi_key
}

search = GoogleSearch(params)
results = search.get_dict()
organic_results = results["organic_results"]

for result in organic_results:
    print(f"Title: {result.get('title', 'N/A')}")
    print(f"Link: {result.get('link', 'N/A')}")
    print("-" * 20)
```
